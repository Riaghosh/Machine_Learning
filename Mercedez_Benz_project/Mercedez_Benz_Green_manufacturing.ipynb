```python
#Check if any nan are present in any column of the data
#Note this part is not required if the previous total nan count in the entire dataset gives 0
##--------We have kept this code here to reuse it for other cases as required
count_nan_cols=0
for col in df_train.columns:
    if (df_train[col].isna().sum())!=0:    
        print(col)
        count_nan_cols+=1
print("# cols containing nan in the training data=",count_nan_cols)

count_nan_cols_test=0
for col_test in df_test.columns:
    if (df_test[col_test].isna().sum())!=0:    
        print(col_test)
        count_nan_cols_test+=1
print("# cols containing nan in the test data=",count_nan_cols_test)
```

```python
#---------------checking the features and cols in the train and test data set-------------
#extracting the cols from the two data frames in two lists
train_cols=list(df_train.columns)
test_cols=list(df_test.columns)
#checking the length of the lists just creted
print("# cols in the train data:",len(train_cols))
print("# cols in the test data:",len(test_cols))

#removing the target variable from the train data set, so that comparing the features becomes easy
train_cols.remove('y')
print("len train_cols removing target var:",len(train_cols))

# getting the unique elements of the lists by converting them to two sets
set(train_cols)
set(test_cols)
#checking the number of unique elements in the sets 
print("#unique features in training data after removing target var:",len(set(train_cols)))
print("#unique features in test data :",len(set(test_cols)))

if (set(train_cols) == set(test_cols)):
    print("The features in the train and the test data set are identical")
else:
    print(("the features in the train and the test data set are NOT identical, further processing required"))
```

```python
##----------------Checking mismatch in categorical featues and values-----------------
##creating dictionaries of categorical values in train and test data set
#----------------For Train Data----------------------
cat_col=[]
for col in df_train.columns:
    if df_train.dtypes[col]=='object':
        cat_col.append(col)

values=np.empty(len(cat_col), dtype=int)
for i in range(len(cat_col)):     
    values[i]= len(df_train[cat_col[i]].unique())

print("cat_col_train:",cat_col)    
print("values_train:",values)
dict_cat_train = dict(zip(cat_col, values))
print("dict_cat_train=",dict_cat_train)

#----------------For test data----------------------

cat_col_=[]
for col_ in df_test.columns:
    if df_test.dtypes[col_]=='object':
        cat_col_.append(col_)

values_=np.empty(len(cat_col_), dtype=int)
for j in range(len(cat_col_)):     
    values_[j]= len(df_test[cat_col_[j]].unique())

print("cat_col_test:",cat_col_)    
print("values_test:",values_)
dict_cat_test = dict(zip(cat_col_, values_))
print("dict_cat_test=",dict_cat_test)

##--------------Checking for any mismatch in the keys and it's values---------------------
mismatch_key=[]
for key in dict_cat_train:
    if key in dict_cat_test.keys():        
        if (dict_cat_train[key]==dict_cat_test[key]):
            continue
        else:
            print(key,dict_cat_train[key],dict_cat_test[key])
            mismatch_key.append(key)
    else:
        print("There's a mismatch in categorical features in train and test data, further preprocessing required")
        break

print("mismatch_key:",mismatch_key)
```

```python
##-----Collect the categorical features in the test set which are not in the train set
## We need to drop these as the model trained on the train set won't learn anything about these features anyway
## So dropping these would improve the computational efficiency of prediction 
s=[]
for key in mismatch_key:
    cat_train_unique=df_train[key].unique()
    cat_test_unique=df_test[key].unique()    
    s.append([ x for x in cat_test_unique if x not in cat_train_unique])
    print("len(df_train[key].unique()):",len(df_train[key].unique()))
    print("len(df_test[key].unique()):",len(df_test[key].unique()))


print("len(s[0]):",len(s[0]))
print("s[0]:",s[0])
print("len(s[1]):",len(s[1]))
print("s[1]:",s[1])
print("len(s[2]):",len(s[2]))
print("s[2]:",s[2])

##-----mismatch_key is the list containing the list of features with mismatch
##------s is a list containing the mismatches corresponding to the mismatched features in the list mismatch_key
## SO creating a dictionary out of it

dict_mismatch = dict(zip(mismatch_key, s))
```
