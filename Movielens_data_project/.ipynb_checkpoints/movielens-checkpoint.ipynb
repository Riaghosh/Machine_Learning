## Feature Engineering:

```python
# splitting the different genres in the column Genre
genre_split=data_movie['Genre'].str.split('|')
genre_split
```

```python
#separate column for each genre category with a one-hot encoding ( 1 and 0) whether or not the movie belongs to that genre.
genre_sep=data_movie['Genre'].str.get_dummies()
genre_sep
```

```python
# Finding out all the unique genres
print(genre_sep.columns)
print(len(genre_sep.columns))
```

```python
# features affecting the ratings of any particular movie
# gender, age, occupation, genre
# we create a dataframe consisting of only those relevant features

Ratings_Data=pd.concat([Master_Data[['Gender', 'Age', 'Occupation','Rating']],genre_sep], axis=1)
# replace Male(M) by 1 and Female(F) by 0 in the Gender column
Ratings_Data['Gender'].replace({'M': 1, 'F': 0 }, inplace=True)
```

```python
Ratings_Data.head(10)
```

```python
# Find the correlation of the other features with Rating to see which is maximum correlated
Ratings_Data1=Ratings_Data[['Age','Occupation','Rating','Gender']]
Ratings_Data1[Ratings_Data1.columns].corr()['Rating']
```

## An appropriate model to predict the movie ratings: Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
X_feature=Ratings_Data1.drop(['Rating'],axis=1)
Y_target=Ratings_Data1['Rating']
```

```python
print(X_feature.shape)
X_feature.head
```

```python
print(Y_target.shape)
Y_target.head
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_feature, Y_target, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train)
```

```python
y_pred = lin_reg.predict(x_test)
```

```python
y_pred 
```

```python
# print the values obtained from different classification metrics 
print('y-intercept: ', lin_reg.intercept_)
print('Beta coefficients: ',lin_reg.coef_)
print('Mean Sq Error  MSE: ',metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Sq Error RMSE:',np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('r2 value: ',metrics.r2_score(y_test, y_pred))
```

```python
# print the first 20 actual and predicted responses
print('actual:    ', y_test.values[0:10])
print('predicted: ', y_pred[0:10])
```
